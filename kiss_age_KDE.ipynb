```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 讀取你的資料檔案（請自行指定檔案路徑）
df = pd.read_excel("DATA_Kiss_count_gender_and_IQ.xlsx")

# 整理欄位
df['Gender'] = df['Gender'].str.strip().str.lower()
df['GenderFlag'] = df['Gender'].apply(lambda x: 1 if x == 'male' else 0)

# 移除缺失值
df = df.dropna(subset=['Kiss Count', 'Age of First Kiss'])

# 繪製 KDE plot
plt.figure(figsize=(8, 5))
sns.kdeplot(
    x=df["Age of First Kiss"],
    y=df["Kiss Count"],
    cmap="Blues",
    fill=True,
    thresh=0.01,     # 控制邊界消失區
    bw_adjust=1      # 平滑程度調整
)
plt.title("KDE: Kiss Count vs. Age of First Kiss", fontsize=14)
plt.xlabel("Age of First Kiss")
plt.ylabel("Kiss Count")

# 儲存圖檔（可選）
plt.savefig("kiss_vs_age_kde.png")
plt.show()
```
