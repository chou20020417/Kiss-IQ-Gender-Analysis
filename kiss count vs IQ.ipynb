```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import pearsonr

df = pd.read_excel("DATA_Kiss_count_gender_and_IQ.xlsx")
df['Gender'] = df['Gender'].str.strip().str.capitalize()
df = df.dropna(subset=['Gender', 'Kiss Count'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

#  p 
r2, p2 = pearsonr(df['IQ'], df['Kiss Count'])

#  Seaborn 
sns.set(style="darkgrid", context="notebook")

# 
plt.figure(figsize=(7, 5))
sns.regplot(
    x='IQ',
    y='Kiss Count',
    data=df,
    scatter_kws={'alpha': 0.6, 'color': '#1f77b4'},  # 
    line_kws={'color': '#d62728', 'linewidth': 2},   # 
)
plt.title('Kiss Count vs IQ', fontsize=14)
plt.xlabel('IQ')
plt.ylabel('Kiss Count')
plt.tight_layout()
plt.savefig("kiss_count_vs_iq.png")
plt.show()
```
