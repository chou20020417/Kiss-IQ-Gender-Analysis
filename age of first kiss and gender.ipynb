```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("DATA_Kiss_count_gender_and_IQ.xlsx")
df['Gender'] = df['Gender'].str.strip().str.lower()
df['Age of First Kiss'] = pd.to_numeric(df['Age of First Kiss'], errors='coerce')
df_clean = df[df['Gender'].isin(['male', 'female'])].dropna(subset=['Age of First Kiss'])

plt.figure(figsize=(8, 5))
sns.histplot(
    data=df_clean,
    x='Age of First Kiss',
    hue='Gender',
    multiple='stack',
    bins=10,
    palette='coolwarm'
)
plt.xlabel("Age of First Kiss")
plt.ylabel("Number of Participants")
plt.grid(True)
plt.tight_layout()
plt.savefig("age_first_kiss_gender.png")
plt.show()
```
