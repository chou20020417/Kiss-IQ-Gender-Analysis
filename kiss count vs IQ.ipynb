```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import pearsonr

df = pd.read_excel("DATA_Kiss_count_gender_and_IQ.xlsx")
df['Gender'] = df['Gender'].str.strip().str.capitalize()
df = df.dropna(subset=['Gender', 'Kiss Count'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 
plt.figure(figsize=(8, 6))

# 
sns.scatterplot(
    data=df,
    x="IQ",
    y="Kiss Count",
    hue="Gender",           # 
    alpha=0.6
)

# 
sns.regplot(
    data=df,
    x="IQ",
    y="Kiss Count",
    scatter=False,
    color='black',
    label='Overall Trend'
)

# 
plt.title("Scatter Plot: Kiss Count vs. IQ by Gender")
plt.xlabel("IQ")
plt.ylabel("Kiss Count")
plt.legend(title="Gender")
plt.grid(True)
plt.tight_layout()

# 
plt.savefig("kiss_iq_gender_scatter.png")
plt.show()
```
